```python
import numpy as np
```

```python
def normalize_vector(x):
    return x / x.sum()
a = np.random.rand(3)
normalize_vector(a)
```

```python
def normalize_last_axis_of_tensor(x):
    return x / x.sum(axis=-1, keepdims=True)
a = np.random.rand(3, 4, 5)
normalize_last_axis_of_tensor(a)
```

```python
def compute_mse(x, y):
    return np.mean((x - y)**2)

a = np.random.rand(3, 4, 5)
b = np.random.rand(3, 4, 5)
compute_mse(a, b)
```

```python
def compute_mse_last_axis(A, b):
    return np.mean((A - b)**2, axis=-1)

X = np.random.rand(3, 4, 5)
y = np.random.rand(5)
compute_mse_last_axis(X, y)
```

```python
def outer_product(a, b):
    return a.reshape(-1, 1) * b.reshape(1, -1)

x = np.random.rand(3)
y = np.random.rand(4)
(outer_product(x, y) == np.outer(x, y)).all()
```

```python
def subtract_constant(A):
    B = A[..., np.newaxis] - [3.5, 2.5]
    return B

a = np.array([10])
sol_a = np.array([[6.5, 7.5]])
print((subtract_constant(a) == sol_a).all())

b = np.array([[5, 3, 0], [0, 2, 6]])
sol_b = np.array(
    [[[ 1.5,  2.5],
    [-0.5,  0.5],
    [-3.5, -2.5]],
    [[-3.5, -2.5],
    [-1.5, -0.5],
    [ 2.5,  3.5]]]
)
print((subtract_constant(b) == sol_b).all())
```
